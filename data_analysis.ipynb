### RNASeq Analysis:  Exploration of biological and technical variation by stratifying gene expression based on expression in bulk RNA

```python
tpm_df = helpers.filter_df(helpers.load_tpms(rsem_dir, sample_map), 
                            genes_only=True,
                            expressed_in_multiple=False)
```

#### Bin bulk RNA expression genes into low, mid, and high

```python
control = 'Total RNA-100pg-2' # Set control sample name
low, mid, high = helpers.get_low_mid_high_genes(tpm_df[control])
control_expressed = tpm_df[tpm_df[control] > 0][[control]]
```

```python
df = pd.concat([low.describe(), mid.describe(), high.describe()], axis=1)
df.columns = ['Low', 'Mid', 'High']
df
```

#### Calculate fraction of genes represented in each sample

```python
expressed_df = helpers.calculate_relative_expression(tpm_df, low, mid, high)
```

```python
expressed_df
```

#### For each sample, plot percentage of bulk RNA represented

```python
cols = ['Low','Mid','High']
control_values = expressed_df[cols].ix[control]
fraction_df = expressed_df[cols].astype(float)/control_values
ax = fraction_df.plot(kind='barh')
ax.set_title('Fraction of %s Genes Expressed' % control);
ax.set_xlabel('Fraction of Genes Expressed');
```

#### For each sample, calculate composition compared to bulk RNA

```python
composition_df = expressed_df.apply(lambda x: x.astype(float)/x.sum(),
                                    axis=1)
cols = ['Low', 'Mid', 'High', 'Novel']
ax = composition_df[cols].plot(kind='barh')
ax.set_title('Composition of Genes Expressed \nRelative to Expression in %s' % control,
             loc='left')
ax.set_xlabel('Fraction of Genes Expressed');
```
