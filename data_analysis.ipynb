### RNASeq Analysis: Quality control based on 3’ bias in cDNA synthesis and sequencing depths across gene body

#### Create bed file for highly expressed genes in bulk sample

```python
def extract_field(fieldname, information):
    try:
        return re.search(fieldname+' "([\w.-]+)"', information).group(1)
    except:
        return "None"
    
high_gtf = gtf_file.replace('.gtf','.high_expressed.test.gtf')

print "Subsetting %s to only highly expressed genes" % (gtf_file)

with open(gtf_file) as IN:
    with open(high_gtf, 'w') as OUT:
        for line in IN:
            if line.startswith('#'):
                OUT.write(line)
            else:
                gene_id = extract_field("gene_id", line)
                if gene_id in high.index:
                    OUT.write(line)
```

```python
high_bed = high_gtf.replace('.gtf','.bed')
print "Converting %s to %s" % (high_gtf, high_bed)
#!perl $gtf_to_bed $high_gtf > $high_bed
```

#### (Optional) Create intron and exon bed files from provided gtf

```python
%%bash -s "$gtf_file" "$exonic_bed" "$intronic_bed" 
#Script to create exonic, intronic, and intergenic beds.
#Adapted from http://crazyhottommy.blogspot.com/2013/05/find-exons-introns-and-intergenic.html

GTF_FILE=$1
EXON_BED=$2
INTRON_BED=$3

#Create Exon Bed
cat $GTF_FILE | awk 'BEGIN{FS="\t";OFS="\t";} $3=="exon" {print $1,$4,$5,$9}' > $EXON_BED

sortBed -i $EXON_BED > temp.bed

mv -f temp.bed $EXON_BED

#Merge Exons Together

mergeBed -i $EXON_BED -c 1,4 -o count,collapse > tmp_merged_exon.bed


#Create Intron Bed
cat $GTF_FILE | awk 'BEGIN{FS="\t";OFS="\t";} $3=="gene" {print $1,$4,$5,$9}' > tmp.bed

sortBed -i tmp.bed > tmp_gene.bed

subtractBed -a tmp_gene.bed -b tmp_merged_exon.bed > $INTRON_BED
rm tmp_merged_exon.bed
rm tmp.bed
rm tmp_gene.bed
```
